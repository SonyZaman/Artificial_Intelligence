## 4. While Loop: ATM PIN System

```python
correct_pin = "1234"
attempts = 0 # 3 times attempt , Maximum thereshold =3
while attempts < 3:
    pin = input("Enter PIN: ")
    if pin == correct_pin:
        print("Access Granted")
        break
    attempts += 1
else:
    print("Card Blocked")

print("New logic start from here")

```

## 5. Mini Project: Smart User Validator

```python
raw_data = "   USER:2025-Ostad-77 | STATUS:active   "
print(raw_data)                      #   USER:2025-Ostad-77 | STATUS:active
clean_data = raw_data.strip()
print(clean_data)                    #USER:2025-Ostad-77 | STATUS:active
clean_data=clean_data.upper()
print(clean_data)                    #USER:2025-OSTAD-77 | STATUS:ACTIVE
sections= clean_data.split("|")
print(sections)                      #['USER:2025-OSTAD-77 ', ' STATUS:ACTIVE']
user_id = sections[0].split(":")[1]
print(user_id)                       #2025-OSTAD-77
status = sections[1].split(":")[1]
print(status)                        #ACTIVE
user_id[:4]                          #2025
```

```python
raw_data = "   USER:2025-Ostad-77 | STATUS:active   "
clean_data = raw_data.strip().upper()
sections = clean_data.split("|")
user_id = sections[0].split(":")[1]
status = sections[1].split(":")[1]
year = user_id[:4]

if status == "ACTIVE" and year == "2025":
    print(f"User {user_id} authorized")
else:
    print("Access denied")
```
