## 3. Break & Continue â€“ Payment Processing Pipeline

These statements give you finer control over your loops.

    break: Stops the loop entirely, even if the loop condition is still true.

    continue: Skips the current iteration and moves to the next item in the sequence.

```python
# break : loop exit
# continue : start from beginning of loop
```

```python

shirts = ["Blue Denim", "White Linen", "Black Cotton"]
pants = ["Khaki Chinos", "Black Jeans"]

for shirts in shirts:
    for pant in pants:
        if pant == "Black Jeans":
            break
        print(f"{shirts} + {pant}")

```

## 5. List Comprehension â€“ Data Transformation




### When, Where, and Why to Use It

List comprehension is a **clean, expressive, and Pythonic** way to create new lists from existing data. It is widely used in **professional Python codebases** for data processing, filtering, and transformation.



### âœ… WHEN to Use List Comprehension

Use list comprehension when you want to:

- Create a **new list** from another list or iterable
- **Transform** each element (e.g., formatting, calculations)
- **Filter** data based on a condition
- Write **short, readable, single-purpose logic**
- Replace a simple `for` loop that only appends values

### Benefits of List Comprehensions

List comprehensions offer several advantages over traditional `for` loops for creating lists:

1.  **Conciseness**: They allow you to create new lists in a single, compact line of code, reducing the overall lines of code and making it easier to read and understand the intent.

2.  **Readability**: Once familiar with the syntax, list comprehensions are often more readable than `for` loops for simple list construction and transformation tasks, as they express the entire operation in a clear, declarative way.

3.  **Performance**: In many cases, list comprehensions can be slightly faster than equivalent `for` loops, especially for larger datasets. This is because they are optimized at the C level in Python's interpreter.

4.  **Reduced Boilerplate**: They eliminate the need for `append()` calls and explicit loop initialization, leading to less boilerplate code.

### A. Single Item Transformation

**Use Case: Cleaning User Input** When users sign up for your website, they often type their names with messy capitalization. You need to standardize them for your database.

```python
raw_usernames = [
    "  john_doe ",
    "SARA99",
    " miKe_tySon ",
    "admin"
]

# --- Using List Comprehension (Recommended) ---
normalized_lc = [
    username.strip().lower()
    for username in raw_usernames
]

# --- Using a Traditional For Loop ---
normalized_loop = []

for username in raw_usernames:
    cleaned = username.strip().lower()
    normalized_loop.append(cleaned)

print("List Comprehension Output:", normalized_lc)
print("For Loop Output:", normalized_loop)
```

### B. Filtering Data

**Use Case: Finding Premium Customers** From a list of order totals, extract only the "High Value" orders (over $500) to send them a VIP discount code.

```python
daily_orders = [120, 550, 45, 1200, 300, 750]
VIP_THRESHOLD = 500

# --- Using List Comprehension (Recommended) ---
vip_orders_lc = [
    order for order in daily_orders
    if order >= VIP_THRESHOLD
]

# --- Using a Traditional For Loop ---
vip_orders_loop = []

for order in daily_orders:
    if order >= VIP_THRESHOLD:
        vip_orders_loop.append(order)

print("VIP Orders (List Comprehension):", vip_orders_lc)
print("VIP Orders (For Loop):", vip_orders_loop)
```

### C. Pair/Tuple Generation (Advanced)

**Use Case: Inventory Value Calculator** You have a list of stock counts and you want to create a list of pairs (tuples) containing the item name and its total value (Quantity Ã— Price).

```python
items = ["Laptop", "Mouse", "Monitor"]
stock_counts = [10, 50, 20]
price_per_unit = 1000

# --- Using List Comprehension ---
inventory_valuation_lc = [(items[i], stock_counts[i] * price_per_unit) for i in range(len(items))]

# --- Using a traditional For Loop ---
inventory_valuation_for = []
for i in range(len(items)):
    inventory_valuation_for.append((items[i], stock_counts[i] * price_per_unit))

print("List Comprehension Result:")
print(inventory_valuation_lc)
print("For Loop Result:")
print(inventory_valuation_for)
```
