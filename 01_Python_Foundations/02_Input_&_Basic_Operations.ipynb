## üßÆ Sample Project: Simple Calculator

```python
# ‚úÖ Calculator Code (Beginner Friendly)

print("Simple Calculator")
print("Operations: +  -  *  /")

# Take inputs
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Choose an operator (+, -, *, /): ") ## *

# Logic
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 == 0:
        result = "Error! Division by zero."
    else:
        result = num1 / num2
else:
    result = "Invalid operator"

# Output
print(f"Result: {result}")

# üß™ Example Run
# Enter first number: 10
# Enter second number: 5
# Choose an operator (+, -, *, /): *
# Result: 50.0
```
