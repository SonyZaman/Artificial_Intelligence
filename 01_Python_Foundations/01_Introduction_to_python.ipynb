## What is Mutable in Python? 

###Mutable means:
 You can change the value of an object after it is created, without creating a new object.

###Simple Definition (Beginner Friendly)

If a data type is mutable, you can:

* Modify it

* Add items to it

* Remove items from it

If it is immutable, you cannot change it once its created.

### Mutable vs Immutable (Quick Idea)

**Think of it like this:**

* Mutable   You can edit it

* Immutable   Read-only, no editing

image.png
### Properties
- Mutable
- Key-value pairs
- Accessed by key, not index


```python
val1={"name":"Tahmid","company":"Ibusinessformula"}
val1["experience"]=3
val1["company"]="brainstation23"
val1.popitem()
val1
```


## G. Sets (`set`)
image.png

### Properties
- Mutable

```python
my_set = {10,1, 2, 2, 3}
my_set.add(4)
my_set.remove(1)

print(my_set)
```


## Set Operations


```python

a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```
